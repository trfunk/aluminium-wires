```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_theme()
```

```python
plastic_strains = []
stresses = []
with open("MISO.dat") as miso:
    for line in miso:
        coeff = line.split(",")
        plastic_strains.append(float(coeff[0]))
        stresses.append(float(coeff[1]))

for stress in stresses:
    print(stress)
```

```python
sns.lineplot(x=plastic_strains, y=stresses)
plt.xlabel(r"plastic strain $|\epsilon_p|$ [-]")
plt.ylabel(r"true stress $|\sigma_t|$ [MPa]")
plt.savefig("miso.pdf")
```
